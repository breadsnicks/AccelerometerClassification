```python
import random
from collections import deque
```

```python
queue = deque()
serviceRate = 0.75
timeslots = 1000000
systemTimes = []
avg_queue_size = 0
```

```python
# adjust arrival rate
arrivalRate = 0.74
```

```python
def singleServerQueueSim():
  queue_length_sum = 0
  for t in range(1, timeslots + 1):
    # check if arrival happened -> record the time it arrived calculate time spent in queue
    if random.random() < arrivalRate:
      queue.append(t)

    if queue and random.random() < serviceRate:
      # Record service time and calculate time in queue by determining the difference
      # between time of service and
      arrivalTime = queue.popleft()
      timeInQueue = t - arrivalTime
      systemTimes.append(timeInQueue)

    queue_length_sum += len(queue)

  avg_queue_size = queue_length_sum / timeslots
  avg_time = sum(systemTimes) / len(systemTimes)
  print(f"Final Queue: {list(queue)}")
  print(f"Average Waiting Time:", avg_time)
  print(f"Average Queue Size:", avg_queue_size)
  avg_queuing_delay = avg_queue_size / arrivalRate
  print(f"Queuing Delay via Little's law:",avg_queuing_delay)
```

```python
singleServerQueueSim()
```
